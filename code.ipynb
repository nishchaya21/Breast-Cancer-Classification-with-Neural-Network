```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Features & targets

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

Training & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

Building Neural Network

```python
# importing libraries
import tensorflow as tf
tf.random.set_seed(2)
from tensorflow import keras
```

```python
# setting up the layers
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(30, activation='relu'),
    # keras.layers.Dense(30, activation='sigmoid'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network
model.fit(X_train_std, Y_train, validation_split=0.1, epochs=30,callbacks=[tf.keras.callbacks.CSVLogger('his.csv')])
```

```python
import pandas
his = pandas.read_csv('his.csv') 
his.head()
```

```python
plt.plot(his.epoch, his.accuracy, label="accuracy")
plt.plot(his.epoch, his.val_accuracy, label="val_accuracy")

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

Accuracy

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
